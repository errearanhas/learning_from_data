```python
import matplotlib.pyplot as plt
import numpy as np
import random
% matplotlib inline
```

```python
pt1=[random.uniform(-1,1),random.uniform(-1,1)]
pt2=[random.uniform(-1,1),random.uniform(-1,1)]

a, b = np.polyfit(pt1, pt2, deg=1)

x_reta = [-1, pt1[0], pt2[0], 1]
y_reta = [(a*x + b) for x in x_reta]

w2=1
w1=(-a)
w0=(-b)

w=[w0,w1,w2]

plt.ylim(-1,1)
plt.xlim(-1,1)

x_acima=[]
x_abaixo=[]
y_acima=[]
y_abaixo=[]

for i in range(500):
    x=random.uniform(-1,1)
    y=random.uniform(-1,1)
    if np.dot((1,x,y),(w)) < 0:
        x_acima.append(x)
        y_acima.append(y)
    else:
        x_abaixo.append(x)
        y_abaixo.append(y)
    
plt.plot(x_reta,y_reta,marker='o',color='b')
plt.plot(x_abaixo,y_abaixo,'ro')
plt.plot(x_acima,y_acima,'bx')
        

print(pt1)
print(pt2)
w=plt.show()
```
